Calculate total marks and percentage
Assign grade for each subject based on marks:

90-100: A+

80-89: A

70-79: B+

60-69: B

50-59: C

40-49: D

Below 40: F (Fail)

Determine overall result (Pass/Fail)


```python
print("--Grade Calculator--")

total_marks = 0
failed_subjects_count = 0
subject_grades = []

for i in range(8):
    try:
        subject_mark = int(input(f"Enter your marks for subject {i+1}: "))

        if not (0 <= subject_mark <= 100):
            print(f"Warning: Marks for subject {i+1} ({subject_mark}) are out of the 0-100 range. Assuming 0 for calculation purposes.")
            subject_mark = 0
            grade = "Invalid"
        elif 90 <= subject_mark <= 100:
            grade = "A+"
        elif 80 <= subject_mark <= 89:
            grade = "A"
        elif 70 <= subject_mark <= 79:
            grade = "B+"
        elif 60 <= subject_mark <= 69:
            grade = "B"
        elif 50 <= subject_mark <= 59:
            grade = "C"
        elif 40 <= subject_mark <= 49:
            grade = "D"
        else:
            grade = "F (Fail)"
            failed_subjects_count += 1

        print(f"Grade for subject {i+1}: {grade}")
        subject_grades.append(grade)
        total_marks += subject_mark

    except ValueError:
        print(f"Invalid input for subject {i+1}. Please enter a numerical value. Assuming 0 marks for this subject.")
        grade = "Invalid (Input Error)"
        subject_grades.append(grade)
        failed_subjects_count += 1
    except Exception as e:
        print(f"An unexpected error occurred for subject {i+1}: {e}")
        grade = "Error"
        subject_grades.append(grade)
        failed_subjects_count += 1



print("\n--- Overall Results ---")

print(f"Total Marks: {total_marks}")
overall_percentage = (total_marks / (8 * 100)) * 100 if 8 > 0 else 0
print(f"Overall Percentage: {overall_percentage:.2f}%")

if failed_subjects_count > 0:
    print("Overall Result: Fail (One or more subjects were failed or had invalid input)")
else:
    print("Overall Result: Pass")

print("Performance Remarks:")
if overall_percentage >= 90:
    print("Outstanding")
elif overall_percentage >= 80:
    print("Excellent")
elif overall_percentage >= 70:
    print("Very Good")
elif overall_percentage >= 60:
    print("Good")
elif overall_percentage >= 50:
    print("Satisfactory")
elif overall_percentage >= 40:
    print("Needs Improvement")
else:
    print("Poor Performance (Below 40% overall)")
```
